```python
import math
import random
import numpy as np
import matplotlib.pyplot as plt

# weight array with 100 random float numbers between 40.0 and 90.0
wt = np.random.uniform(40, 90, size=100) 

# height array with 100 random integers between 140 and 200
ht = np.random.randint(140, 200, size=100)

# calculate bmi
bmi = wt/(ht*ht)*10000

# first 10 elements of the bmi
for i in range(0,10):
    print(bmi[i])
```

```python
# store the number of each bmi level
underweight = 0
healthy = 0
overweight = 0
obese = 0

b = [] # store the bmi level in numeric

# store height for each bmi level 
underweightH = []
healthyH = []
overweightH = []
obeseH = []

# store weight for each bmi level
underweightW = []
healthyW = []
overweightW = []
obeseW = []

for i in range(0, 100):
    if (bmi[i] < 18.5) : # underweight
        b.append(0)
        underweightH.extend([ht[i]/100])
        underweightW.extend([wt[i]])
        underweight += 1
    elif (bmi[i] >= 18.5 and bmi[i] < 25): # healthy
        b.append(1)
        healthyH.extend([ht[i]/100])
        healthyW.extend([wt[i]])
        healthy += 1 
    elif (bmi[i] >= 25.0 and bmi[i] < 30): # overweight
        b.append(2)
        overweightH.extend([ht[i]/100])
        overweightW.extend([wt[i]])
        overweight += 1 
    else : # obese
        b.append(3)
        obeseH.extend([ht[i]/100])
        obeseW.extend([wt[i]])
        obese += 1 
```

```python
# box plot

# height
plotDataH = [underweightH, healthyH, overweightH, obeseH]
plt.boxplot(plotDataH)
plt.xticks([1,2,3,4], ['underweight', 'healthy', 'overweight', 'obese'])
plt.title("<Box plot>")
plt.xlabel("BMI status")
plt.ylabel("Height (m)")
plt.show()

# weight
plotDataW = [underweightW, healthyW,overweightW, obeseW]
plt.boxplot(plotDataW)
plt.xticks([1,2,3,4], ['underweight', 'healthy', 'overweight', 'obese'])
plt.xlabel("BMI status")
plt.ylabel("Weight (kg)")
plt.show()
```

```python
# Histogram
ys, xs, patches = plt.hist(b, bins=4, rwidth=0.8)
plt.title('<Histogram>')
plt.xlabel('BMI status')
plt.ylabel('Number of students')

plt.xticks([(xs[i]+xs[i+1])/2 for i in range(0, len(xs)-1)], ["underweight", "healthy", "overweight", "obese"])
plt.yticks([0,10,20,30,40])
plt.show()
```

```python
# pie chart
langs = ['Underweight', 'Healthy', 'Overweight', 'Obese']
status = [underweight, healthy, overweight, obese]

plt.pie(status, labels = langs, autopct='%1.2f%%')
plt.axis('equal')
plt.title('<pie chart>')
plt.show()
```

```python
# scatter plot
plt.scatter(ht, wt, color = 'g')
plt.xlabel('height')
plt.ylabel('weight')
plt.title('<scatter plot>')
```
